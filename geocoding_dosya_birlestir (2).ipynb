```python
import os
import pandas as pd

# Klasör yolu
folder_path = "/Users/canyonder/Desktop/wrong matches"

# Klasördeki tüm CSV dosyalarını bul
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Boş bir liste oluştur
dataframes = []

# CSV dosyalarını oku ve filtrele
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    try:
        df = pd.read_csv(file_path)  # CSV dosyasını oku
        df = df[df['Match'].astype(str) == 'No Info']  # "Match" sütunu "no info" olanları filtrele
        if not df.empty:
            df['Dosya Adı'] = file  # Kaynak dosyayı takip etmek için sütun ekleyelim
            dataframes.append(df)
    except Exception as e:
        print(f"❌ {file} okunamadı veya 'Match' sütunu yok, atlanıyor... Hata: {e}")

# Eğer en az bir dosya okunmuşsa birleştir
if dataframes:
    merged_df = pd.concat(dataframes, ignore_index=True)
    output_file = os.path.join(folder_path, "birlesik_no_info.csv")
    merged_df.to_csv(output_file, index=False)
    print(f"✅ 'no info' içeren satırlar {output_file} olarak birleştirildi!")
else:
    print("⚠ Hiçbir uygun veri bulunamadı, birleşik dosya oluşturulmadı.")
```

```python
import pandas as pd
import os

# Dosya yolu
file_path = "/Users/canyonder/Desktop/wrong matches/birlesik_dosya.csv"

# Yeni dosyanın adı
output_file = "/Users/canyonder/Desktop/wrong matches/birlesik_temiz.csv"

try:
    # CSV dosyasını oku
    df = pd.read_csv(file_path)

    # "Match" sütununda "No Info" ve "Wrong" içeren satırları filtrele
    df_cleaned = df[~df['Match'].astype(str).str.lower().isin(['no info', 'wrong'])]

    # Temizlenmiş veriyi yeni dosyaya kaydet
    df_cleaned.to_csv(output_file, index=False)

    print(f"✅ 'No Info' ve 'Wrong' olan satırlar çıkarıldı. Yeni dosya: {output_file}")

except Exception as e:
    print(f"❌ Hata oluştu: {e}")
```

```python
import pandas as pd

# Excel dosyasını okuma
file_path = "/Users/canyonder/Desktop/tam_listeler/birlesik_wrong_matches.xlsx"

# Hatalı satırları saklamak için bir liste oluşturuyoruz
bad_lines = []

try:
    # Excel dosyasını okuma
    df = pd.read_excel(file_path, engine='openpyxl')
    print(f"Dosya başarıyla okundu: {file_path}")
    
    # Sütunları yazdırarak mevcut sütun adlarını kontrol etme
    print("Mevcut Sütunlar:", df.columns)

    # 'UnitName', 'CityName' ve 'DistrictName' sütunlarını birleştirme
    if 'UnitName' in df.columns and 'CityName' in df.columns and 'DistrictName' in df.columns:
        df['Combined'] = df['UnitName'] + " " + df['CityName'] + " " + df['DistrictName']
    else:
        print("Eksik sütunlar: UnitName, CityName veya DistrictName")

    # Sonuçları görmek için ilk birkaç satırı yazdırma
    print(df[['UnitName', 'CityName', 'DistrictName', 'Combined']].head())

    # Excel dosyasına kaydetme
    output_path = "/Users/canyonder/Desktop/tam_listeler/updated_wrong_matches.xlsx"
    df.to_excel(output_path, index=False, engine='openpyxl')  # 'openpyxl' Excel için kullanılacak motor
    
    print(f"Veriler başarıyla Excel dosyasına kaydedildi: {output_path}")
    
except Exception as e:
    print(f"Hata oluştu: {e}")

# Hatalı satırları gösterme
if bad_lines:
    print("\nHatalı satırlar:")
    for line in bad_lines:
        print(line)
else:
    print("\nHatalı satır bulunmadı.")
```
