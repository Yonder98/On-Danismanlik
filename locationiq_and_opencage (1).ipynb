```python
import pandas as pd

# Verilen dosya yollarını kullanarak işlemi gerçekleştireceğiz
input_file_path = "/Users/canyonder/Desktop/birlesik_missing_coordinates100.xlsx"
output_file_path = "/Users/canyonder/Desktop/tam_listeler/birlesik_missing_coordinates1.xlsx"
final_output_file_path = "/Users/canyonder/Desktop/tam_listeler/final_output.xlsx"  # 3. dosya yolu


# Excel dosyasını yükle
df_input = pd.read_excel(input_file_path)

# 'Corrected Latitude' ve 'Corrected Longitude' sütunlarında veri bulunan satırları seç
filtered_rows = df_input.dropna(subset=['Corrected Latitude', 'Corrected Longitude'])

# Çıktı dosyasını yükle
df_output = pd.read_excel(output_file_path)

# Eşleşen 'UnitId' değerlerine göre veri ekleyelim
for index, row in filtered_rows.iterrows():
    unit_id = row['UnitId']  # 'UnitId'yi alıyoruz
    
    # 'df_output' dosyasındaki eşleşen 'UnitId' değerine sahip satırı buluyoruz
    matching_row_index = df_output[df_output['UnitId'] == unit_id].index

    if not matching_row_index.empty:
        # Eşleşen satırda 'Corrected Latitude' ve 'Corrected Longitude' verilerini güncelliyoruz
        df_output.loc[matching_row_index[0], 'Corrected Latitude'] = row['Corrected Latitude']
        df_output.loc[matching_row_index[0], 'Corrected Longitude'] = row['Corrected Longitude']

# Yeni dosyayı 3. bir dosyaya kaydedelim
df_output.to_excel(final_output_file_path, index=False)

final_output_file_path  # Bu satırda 3. dosya yolunu döndürüyoruz
```
