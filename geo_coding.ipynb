```python
# Geopy kütüphanesinden Nominatim modülünü içe aktarma  
# Geopy'nin hız sınırlayıcısını (RateLimiter) içe aktarma  
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
# Nominatim coğrafi kodlayıcıyı kendi sunucumuz üzerinden başlatma
geolocator = Nominatim(user_agent='monument_geocoder', domain='nominatim.enigmasolver.com:8082' ,  scheme='http')
```

```python
# Coğrafi kodlama işlemlerini hız sınırlayıcı ile oluşturur  
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=0)  # min_delay_seconds=0 yaparak bekleme süresini kaldır 
```

```python
# Koordinatları almak için fonksiyon
def get_coordinates(location):
    try:
        location_data = geocode(location)
        if location_data:
            return location_data.latitude, location_data.longitude
        else:
            return None, None  # Eğer koordinatlar bulunmazsa
    except Exception as e:
        print(f"Hata oluştu: {e}")
        return None, None  # Hata durumunda None döndürülür

# Combined sütunundaki her satır için koordinatları alma ve işlem durumunu yazdırma
total_rows = len(df)
for index, row in df.iterrows():
    location = row['Combined']
    latitude, longitude = get_coordinates(location)
    df.at[index, 'Latitude'] = latitude
    df.at[index, 'Longitude'] = longitude
    
    # Durum mesajı
    print(f"{index + 1}/{total_rows} - {location}: Latitude: {latitude}, Longitude: {longitude}")

# Sonuçları görmek için ilk birkaç satırı yazdırma
print("\nİlk birkaç satır:")
print(df[['UnitName', 'CityName', 'DistrictName', 'Combined', 'Latitude', 'Longitude']].head())

# Yeni verileri Excel dosyasına kaydetme
output_path = "/Users/canyonder/Desktop/tam_listeler/updated_wrong_matches_with_coordinates.xlsx"
df.to_excel(output_path, index=False, engine='openpyxl')  # 'openpyxl' Excel için kullanılacak motor

print(f"\nVeriler başarıyla Excel dosyasına kaydedildi: {output_path}")
```

```python
# Eksik koordinatları filtreleme
missing_coords = df[df['Latitude'].isna() | df['Longitude'].isna()]

# Eksik koordinatları ekrana yazdırma
print("\nEksik koordinatları olan satırlar:")
print(missing_coords[['UnitName', 'CityName', 'DistrictName', 'Combined']])

# Eksik koordinatları ayrı bir Excel dosyasına kaydetme
missing_output_path = "/Users/canyonder/Desktop/tam_listeler/missing_coordinates.xlsx"
missing_coords.to_excel(missing_output_path, index=False, engine='openpyxl')

print(f"\nEksik koordinatları içeren dosya kaydedildi: {missing_output_path}")
```
