```python
# Jupyter Notebook için asyncio hatasını önler
nest_asyncio.apply()

# Plaka kodu - şehir eşleşmesi
plaka_to_city = {
    1: "Adana", 2: "Adiyaman", 3: "Afyonkarahisar", 4: "Ağri", 5: "Amasya", 6: "Ankara",
    7: "Antalya", 8: "Artvin", 9: "Aydin", 10: "Balikesir", 11: "Bilecik", 12: "Bingöl",
    13: "Bitlis", 14: "Bolu", 15: "Burdur", 16: "Bursa", 17: "Çanakkale", 18: "Çankiri",
    19: "Çorum", 20: "Denizli", 21: "Diyarbakir", 22: "Edirne", 23: "Elaziğ", 24: "Erzincan",
    25: "Erzurum", 26: "Eskişehir", 27: "Gaziantep", 28: "Giresun", 29: "Gümüşhane",
    30: "Hakkari", 31: "Hatay", 32: "Isparta", 33: "Mersin", 34: "İstanbul"
}

# Kullanıcıdan plaka kodunu alma
while True:
    try:
        plaka_kodu = int(input("Lütfen bir plaka kodu girin (örn: 34): "))
        if plaka_kodu in plaka_to_city:
            break
        print("Geçersiz plaka kodu! Lütfen doğru bir değer girin.")
    except ValueError:
        print("Lütfen sadece sayı girin.")

secilen_sehir = plaka_to_city[plaka_kodu]
print(f"Seçilen şehir: {secilen_sehir}")

# CSV dosyasını oku
file_path =  "/Users/canyonder/Desktop/unitsWithCitiesAndCoordinates1.csv"
df = pd.read_csv(file_path)
# Seçilen şehre ait satırları filtrele
df_filtered = df[df["CityName"] == secilen_sehir]

if df_filtered.empty:
    print("Bu şehir için veri bulunamadı.")
    exit()

# Nominatim API'den veri alma fonksiyonu
async def get_nominatim_data(session, lat, lon):
    url = f"https://nominatim.openstreetmap.org/reverse?format=json&lat={lat}&lon={lon}"
    headers = {'User-Agent': 'Mozilla/5.0'}

    for _ in range(3):  # Maksimum 3 deneme hakkı
        try:
            async with session.get(url, headers=headers) as response:
                if response.status == 200:
                    data = await response.json()
                    
                    # Eğer API'den boş veya yanlış veri gelirse
                    if not isinstance(data, dict):  
                        return "Unknown"

                    address = data.get("address", {})
                    
                    # Öncelikle 'city', 'town', 'village' gibi alanları kontrol et
                    city = (address.get("city") or 
                            address.get("town") or 
                            address.get("village") or 
                            address.get("state") or 
                            address.get("region", "Unknown"))
                    
                    return city
        except Exception as e:
            print(f"API hatası: {e}")
            await asyncio.sleep(2)  # Hata olursa 2 saniye bekle ve tekrar dene

    return "Unknown"


# Tüm koordinatlar için asenkron istek atan fonksiyon
async def fetch_all_data(df):
    async with aiohttp.ClientSession() as session:
        tasks = [get_nominatim_data(session, row["Latitude"], row["Longitude"]) for _, row in df.iterrows()]
        return await asyncio.gather(*tasks)  # Paralel olarak çalıştır

# Asenkron fonksiyonu çalıştır ve API’den sonuçları al
results = asyncio.run(fetch_all_data(df_filtered))

# Gelen şehir verisini DataFrame’e ekle
df_filtered["Nominatim_City"] = results

# Şehir adı eşleşmelerini kontrol et
df_filtered["Match"] = df_filtered["CityName"] == df_filtered["Nominatim_City"]

# Yanlış eşleşenleri filtrele
incorrect_matches = df_filtered[df_filtered["Match"] == False]

# İstatistikleri ekrana yazdır
total = len(df_filtered)
incorrect_count = len(incorrect_matches)
correct_count = total - incorrect_count
accuracy = (correct_count / total) * 100 if total > 0 else 0

print(f"\n📊 Toplam kayıt: {total}")
print(f"✅ Doğru eşleşmeler: {correct_count}")
print(f"❌ Yanlış eşleşmeler: {incorrect_count}")
print(f"📈 Doğruluk oranı: %{accuracy:.2f}")

# Sonucu Excel dosyasına kaydet
output_file = f"incorrect_city_matches_{secilen_sehir}.xlsx"
incorrect_matches.to_excel(output_file, index=False)

print(f"\nYanlış eşleşmeler {output_file} dosyasına kaydedildi.")
```
