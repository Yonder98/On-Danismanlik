```python
import pandas as pd
from geopy.geocoders import Nominatim
import nest_asyncio
from unidecode import unidecode  # TÃ¼rkÃ§e karakter dÃ¼zeltme

# Jupyter Notebook iÃ§in asyncio hatasÄ±nÄ± Ã¶nler
nest_asyncio.apply()

# Yeni sÃ¼tunlar ekleyerek sonuÃ§larÄ± saklamak iÃ§in hazÄ±rlÄ±k yap
df_filtered.loc[:, "DetectedCity"] = "No Info"
df_filtered.loc[:, "Match"] = "No Info"

# YanlÄ±ÅŸ eÅŸleÅŸmeleri saklamak iÃ§in boÅŸ bir DataFrame oluÅŸtur
wrong_matches = pd.DataFrame(columns=df_filtered.columns)

# Plaka kodu - ÅŸehir eÅŸleÅŸmesi
plaka_to_city = {
    1: "Adana", 2: "Adiyaman", 3: "Afyonkarahisar", 4: "AÄŸri", 5: "Amasya", 6: "Ankara",
    7: "Antalya", 8: "Artvin", 9: "Aydin", 10: "Balikesir", 11: "Bilecik", 12: "BingÃ¶l",
    13: "Bitlis", 14: "Bolu", 15: "Burdur", 16: "Bursa", 17: "Ã‡anakkale", 18: "Ã‡ankiri",
    19: "Ã‡orum", 20: "Denizli", 21: "Diyarbakir", 22: "Edirne", 23: "ElaziÄŸ", 24: "Erzincan",
    25: "Erzurum", 26: "EskiÅŸehir", 27: "Gaziantep", 28: "Giresun", 29: "GÃ¼mÃ¼ÅŸhane",
    30: "Hakkari", 31: "Hatay", 32: "Isparta", 33: "Mersin", 34: "Ä°stanbul", 35: "Ä°zmir",
    36: "Kars", 37: "Kastamonu", 38: "Kayseri", 39: "Kirklareli", 40: "KirÅŸehir",
    41: "Kocaeli", 42: "Konya", 43: "KÃ¼tahya", 44: "Malatya", 45: "Manisa", 46: "KahramanmaraÅŸ",
    47: "Mardin", 48: "MuÄŸla", 49: "MuÅŸ", 50: "NevÅŸehir", 51: "NiÄŸde", 52: "Ordu",
    53: "Rize", 54: "Sakarya", 55: "Samsun", 56: "Siirt", 57: "Sinop", 58: "Sivas",
    59: "TekirdaÄŸ", 60: "Tokat", 61: "Trabzon", 62: "Tunceli", 63: "Åanliurfa", 64: "UÅŸak",
    65: "Van", 66: "Yozgat", 67: "Zonguldak", 68: "Aksaray", 69: "Bayburt", 70: "Karaman",
    71: "Kirikkale", 72: "Batman", 73: "Åirnak", 74: "BartÄ±n", 75: "Ardahan", 76: "IÄŸdir",
    77: "Yalova", 78: "KarabÃ¼k", 79: "Kilis", 80: "Osmaniye", 81: "DÃ¼zce"
}

# KullanÄ±cÄ±dan plaka kodunu alma
while True:
    try:
        plaka_kodu = int(input("LÃ¼tfen bir plaka kodu girin (Ã¶rn: 34): "))
        if plaka_kodu in plaka_to_city:
            break
        print("GeÃ§ersiz plaka kodu! LÃ¼tfen doÄŸru bir deÄŸer girin.")
    except ValueError:
        print("LÃ¼tfen sadece sayÄ± girin.")

secilen_sehir = plaka_to_city[plaka_kodu]
print(f"SeÃ§ilen ÅŸehir: {secilen_sehir}")


# Åehir adlarÄ±nÄ± kÃ¼Ã§Ã¼k harfe ve TÃ¼rkÃ§e karakterlerden arÄ±ndÄ±rmaya yardÄ±mcÄ± fonksiyon
def normalize_city_name(city_name):
    # KÃ¼Ã§Ã¼k harfe dÃ¶nÃ¼ÅŸtÃ¼rme ve TÃ¼rkÃ§e karakterleri dÃ¼zeltme
    city_name = unidecode(city_name.lower())
    
    # EÄŸer ÅŸehir adÄ± 'merkez' iÃ§eriyorsa, sadece ilk kÄ±smÄ± al
    if "merkez" in city_name:
        city_name = city_name.split()[0]  # Åehir adÄ±nÄ± ilk kÄ±sÄ±mla sÄ±nÄ±rla
    
    return city_name


# SatÄ±rlarÄ± tek tek kontrol et
for index, row in df_filtered.iterrows():
    lat, lon = row["Latitude"], row["Longitude"]
    city_name = row["CityName"]

    try:
        # Nominatim API'den koordinata gÃ¶re ÅŸehir bilgisini al
        location = geolocator.reverse((lat, lon), exactly_one=True)

        if location and "address" in location.raw:
            detected_province = location.raw["address"].get("province", "")
            detected_city = location.raw["address"].get("city", "")
            detected_state = location.raw["address"].get("state", "")
            detected_town = location.raw["address"].get("town", "")
            display_name = location.raw.get("display_name", "")

            # EÄŸer 'province' bilgisi yoksa, diÄŸer bilgileri kullan
            detected_province = detected_province or detected_city or detected_state or detected_town

            if detected_province:
                df_filtered.at[index, "DetectedCity"] = detected_province
            else:
                df_filtered.at[index, "DetectedCity"] = "No Info"

            # Åehir eÅŸleÅŸmesini kontrol et
            detected_province_normalized = normalize_city_name(detected_province)
            city_name_normalized = normalize_city_name(city_name)
            display_name_normalized = normalize_city_name(display_name)

            # EÄŸer town bilgisi varsa
            if detected_town:
                town_normalized = normalize_city_name(detected_town)
                # Hem town hem de city_name display_name iÃ§inde geÃ§iyor mu?
                if (town_normalized in display_name_normalized) and (city_name_normalized in display_name_normalized):
                    df_filtered.at[index, "Match"] = "Correct"
                else:
                    df_filtered.at[index, "Match"] = "Wrong"
                    # YanlÄ±ÅŸ eÅŸleÅŸen satÄ±rlarÄ± ayrÄ± bir DataFrame'e ekle
                    df_non_empty = df.loc[[index]].dropna(axis=1, how='all')
                    wrong_matches = pd.concat([wrong_matches, df_non_empty], ignore_index=True)
            else:
                # EÄŸer town bilgisi yoksa, normal ÅŸehri kontrol et
                if detected_province_normalized == city_name_normalized:
                    df_filtered.at[index, "Match"] = "Correct"
                else:
                    df_filtered.at[index, "Match"] = "Wrong"
                    # YanlÄ±ÅŸ eÅŸleÅŸen satÄ±rlarÄ± ayrÄ± bir DataFrame'e ekle
                    df_non_empty = df.loc[[index]].dropna(axis=1, how='all')
                    wrong_matches = pd.concat([wrong_matches, df_non_empty], ignore_index=True)

        else:
            df_filtered.at[index, "Match"] = "No Info"
            df_filtered.at[index, "DetectedCity"] = "No Info"

    except Exception as e:
        df_filtered.at[index, "Match"] = f"Error: {e}"

    # Ä°ÅŸlem durumunu ekrana yazdÄ±r
    print(f"Row {index}:")
    print(f"Checked: {row['UnitId']} - {row['UnitName']} - {row['CityName']}")
    print(f"Found: {df_filtered.at[index, 'DetectedCity']}")
    print(f"Match: {df_filtered.at[index, 'Match']}")
    print("-" * 40)


# HatalÄ± kayÄ±t sayÄ±sÄ±nÄ± hesapla
num_wrong_matches = len(wrong_matches)

# ğŸ“Š **Ä°statistikleri Hesapla**
total_records = len(df_filtered)
correct_matches = total_records - num_wrong_matches
accuracy = (correct_matches / total_records) * 100 if total_records > 0 else 0

# ğŸ“ **Ä°statistikleri Ekrana YazdÄ±r**
print("\nğŸ“Š Ä°statistikler:")
print(f"ğŸ”¹ Toplam kayÄ±t: {total_records}")
print(f"âœ… DoÄŸru eÅŸleÅŸmeler: {correct_matches}")
print(f"âŒ YanlÄ±ÅŸ eÅŸleÅŸmeler: {num_wrong_matches}")
print(f"ğŸ“ˆ DoÄŸruluk oranÄ±: %{accuracy:.2f}")

# HatalÄ± kayÄ±tlarÄ± Excel olarak kaydet (MasaÃ¼stÃ¼ne)
wrong_matches.to_excel(r"/Users/canyonder/Desktop/wrong_matches.xlsx", index=False)

# GÃ¼ncellenmiÅŸ veriyi kaydet (MasaÃ¼stÃ¼ne)
df_filtered.to_csv("/Users/canyonder/Desktop/checked_coordinates.csv", index=False)

print("Kontrol tamamlandÄ±!")
print(f"YanlÄ±ÅŸ eÅŸleÅŸmeler 'wrong_matches.xlsx' dosyasÄ±na kaydedildi. Toplam hatalÄ± kayÄ±t sayÄ±sÄ±: {num_wrong_matches}")
```
