```python
import pandas as pd
from geopy.geocoders import Nominatim
import nest_asyncio
from unidecode import unidecode  # Türkçe karakter düzeltme

# Jupyter Notebook için asyncio hatasını önler
nest_asyncio.apply()

# Yeni sütunlar ekleyerek sonuçları saklamak için hazırlık yap
df_filtered.loc[:, "DetectedCity"] = "No Info"
df_filtered.loc[:, "Match"] = "No Info"

# Yanlış eşleşmeleri saklamak için boş bir DataFrame oluştur
wrong_matches = pd.DataFrame(columns=df_filtered.columns)

# Plaka kodu - şehir eşleşmesi
plaka_to_city = {
    1: "Adana", 2: "Adiyaman", 3: "Afyonkarahisar", 4: "Ağri", 5: "Amasya", 6: "Ankara",
    7: "Antalya", 8: "Artvin", 9: "Aydin", 10: "Balikesir", 11: "Bilecik", 12: "Bingöl",
    13: "Bitlis", 14: "Bolu", 15: "Burdur", 16: "Bursa", 17: "Çanakkale", 18: "Çankiri",
    19: "Çorum", 20: "Denizli", 21: "Diyarbakir", 22: "Edirne", 23: "Elaziğ", 24: "Erzincan",
    25: "Erzurum", 26: "Eskişehir", 27: "Gaziantep", 28: "Giresun", 29: "Gümüşhane",
    30: "Hakkari", 31: "Hatay", 32: "Isparta", 33: "Mersin", 34: "İstanbul", 35: "İzmir",
    36: "Kars", 37: "Kastamonu", 38: "Kayseri", 39: "Kirklareli", 40: "Kirşehir",
    41: "Kocaeli", 42: "Konya", 43: "Kütahya", 44: "Malatya", 45: "Manisa", 46: "Kahramanmaraş",
    47: "Mardin", 48: "Muğla", 49: "Muş", 50: "Nevşehir", 51: "Niğde", 52: "Ordu",
    53: "Rize", 54: "Sakarya", 55: "Samsun", 56: "Siirt", 57: "Sinop", 58: "Sivas",
    59: "Tekirdağ", 60: "Tokat", 61: "Trabzon", 62: "Tunceli", 63: "Şanliurfa", 64: "Uşak",
    65: "Van", 66: "Yozgat", 67: "Zonguldak", 68: "Aksaray", 69: "Bayburt", 70: "Karaman",
    71: "Kirikkale", 72: "Batman", 73: "Şirnak", 74: "Bartın", 75: "Ardahan", 76: "Iğdir",
    77: "Yalova", 78: "Karabük", 79: "Kilis", 80: "Osmaniye", 81: "Düzce"
}

# Kullanıcıdan plaka kodunu alma
while True:
    try:
        plaka_kodu = int(input("Lütfen bir plaka kodu girin (örn: 34): "))
        if plaka_kodu in plaka_to_city:
            break
        print("Geçersiz plaka kodu! Lütfen doğru bir değer girin.")
    except ValueError:
        print("Lütfen sadece sayı girin.")

secilen_sehir = plaka_to_city[plaka_kodu]
print(f"Seçilen şehir: {secilen_sehir}")


# Şehir adlarını küçük harfe ve Türkçe karakterlerden arındırmaya yardımcı fonksiyon
def normalize_city_name(city_name):
    # Küçük harfe dönüştürme ve Türkçe karakterleri düzeltme
    city_name = unidecode(city_name.lower())
    
    # Eğer şehir adı 'merkez' içeriyorsa, sadece ilk kısmı al
    if "merkez" in city_name:
        city_name = city_name.split()[0]  # Şehir adını ilk kısımla sınırla
    
    return city_name


# Satırları tek tek kontrol et
for index, row in df_filtered.iterrows():
    lat, lon = row["Latitude"], row["Longitude"]
    city_name = row["CityName"]

    try:
        # Nominatim API'den koordinata göre şehir bilgisini al
        location = geolocator.reverse((lat, lon), exactly_one=True)

        if location and "address" in location.raw:
            detected_province = location.raw["address"].get("province", "")
            detected_city = location.raw["address"].get("city", "")
            detected_state = location.raw["address"].get("state", "")
            detected_town = location.raw["address"].get("town", "")
            display_name = location.raw.get("display_name", "")

            # Eğer 'province' bilgisi yoksa, diğer bilgileri kullan
            detected_province = detected_province or detected_city or detected_state or detected_town

            if detected_province:
                df_filtered.at[index, "DetectedCity"] = detected_province
            else:
                df_filtered.at[index, "DetectedCity"] = "No Info"

            # Şehir eşleşmesini kontrol et
            detected_province_normalized = normalize_city_name(detected_province)
            city_name_normalized = normalize_city_name(city_name)
            display_name_normalized = normalize_city_name(display_name)

            # Eğer town bilgisi varsa
            if detected_town:
                town_normalized = normalize_city_name(detected_town)
                # Hem town hem de city_name display_name içinde geçiyor mu?
                if (town_normalized in display_name_normalized) and (city_name_normalized in display_name_normalized):
                    df_filtered.at[index, "Match"] = "Correct"
                else:
                    df_filtered.at[index, "Match"] = "Wrong"
                    # Yanlış eşleşen satırları ayrı bir DataFrame'e ekle
                    df_non_empty = df.loc[[index]].dropna(axis=1, how='all')
                    wrong_matches = pd.concat([wrong_matches, df_non_empty], ignore_index=True)
            else:
                # Eğer town bilgisi yoksa, normal şehri kontrol et
                if detected_province_normalized == city_name_normalized:
                    df_filtered.at[index, "Match"] = "Correct"
                else:
                    df_filtered.at[index, "Match"] = "Wrong"
                    # Yanlış eşleşen satırları ayrı bir DataFrame'e ekle
                    df_non_empty = df.loc[[index]].dropna(axis=1, how='all')
                    wrong_matches = pd.concat([wrong_matches, df_non_empty], ignore_index=True)

        else:
            df_filtered.at[index, "Match"] = "No Info"
            df_filtered.at[index, "DetectedCity"] = "No Info"

    except Exception as e:
        df_filtered.at[index, "Match"] = f"Error: {e}"

    # İşlem durumunu ekrana yazdır
    print(f"Row {index}:")
    print(f"Checked: {row['UnitId']} - {row['UnitName']} - {row['CityName']}")
    print(f"Found: {df_filtered.at[index, 'DetectedCity']}")
    print(f"Match: {df_filtered.at[index, 'Match']}")
    print("-" * 40)


# Hatalı kayıt sayısını hesapla
num_wrong_matches = len(wrong_matches)

# 📊 **İstatistikleri Hesapla**
total_records = len(df_filtered)
correct_matches = total_records - num_wrong_matches
accuracy = (correct_matches / total_records) * 100 if total_records > 0 else 0

# 📝 **İstatistikleri Ekrana Yazdır**
print("\n📊 İstatistikler:")
print(f"🔹 Toplam kayıt: {total_records}")
print(f"✅ Doğru eşleşmeler: {correct_matches}")
print(f"❌ Yanlış eşleşmeler: {num_wrong_matches}")
print(f"📈 Doğruluk oranı: %{accuracy:.2f}")

# Hatalı kayıtları Excel olarak kaydet (Masaüstüne)
wrong_matches.to_excel(r"/Users/canyonder/Desktop/wrong_matches.xlsx", index=False)

# Güncellenmiş veriyi kaydet (Masaüstüne)
df_filtered.to_csv("/Users/canyonder/Desktop/checked_coordinates.csv", index=False)

print("Kontrol tamamlandı!")
print(f"Yanlış eşleşmeler 'wrong_matches.xlsx' dosyasına kaydedildi. Toplam hatalı kayıt sayısı: {num_wrong_matches}")
```
